```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.arange(1,10)
y = x*5
```

```python
plt.plot(x, y);
```

```python
x = np.arange(1,10)
y = x^4
```

```python
plt.plot(x, y, 'r')
plt.show();
```

```python
plt.plot(x, y, 'b--')
plt.show()
```

```python
plt.plot(x, y, 'c')
plt.title('matplotlib sample')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.show()
```

```python
plt.plot(x, y, 'b')
plt.xlim(2, 7)
plt.ylim(2, 10)
plt.title('matplotlib sample')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.show()
```

```python
x = np.arange(1, 10, 0.1)
y = x*0.2
y2 = np.sin(x)
```

```python
plt.plot(x, y2, y)
plt.xlim(0, 10)
plt.ylim(-1, 2)
plt.title('matplotlib sample')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```
