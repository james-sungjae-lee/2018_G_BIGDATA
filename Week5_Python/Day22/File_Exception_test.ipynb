```python
# Ex1

# 파일 'stockcode.txt'을 한번에 읽어서 출력하시오.

file = open('stockcode.txt', 'r')
data = file.read()
print(data)
file.close()
```

```python
# Ex2

# 'stockcode.txt' 파일을 while과 readline()을 이용해서 한줄씩 읽어서 아래와 같이 출력하시오

f = open('stockcode.txt', 'r')

line_num = 1
line = f.readline()

while line:
    print('{} {}'.format(line_num, line), end = '')
    line = f.readline()
    line_num += 1
    
f.close()
```

```python
# Ex3

# 'stockcode.txt' 파일을 for와 readlines()을 이용해서 아래와 같이 출력하시오

file = open('stockcode.txt')
file_read = file.readlines()

for index, line in enumerate(file_read):
    print(index, line)
    
file.close()
```

```python
# Ex5

# 파일 'stockcode.txt' 내용을 'stockcode_copy.txt'에 복사하시오.

file1 = open('stockcode.txt', 'r')
file2 = open('stockcode_copy.txt', 'w')

data = file1.read()
file2.write(data)

file1.close()
file2.close()
```

```python
# Ex6

# 파일 'python.jpg' 을 'python_copy.jpg'로 복사하시오.
# 단, file_size = 1024byte씩 읽어서

bufsize = 1024
f = open('python.jpg', 'rb')
h = open('python_copy.jpg', 'wb')

data = f.read(bufsize)
while data:
    h.write(data)
    data = f.read(bufsize)

f.close()
h.close()
```
