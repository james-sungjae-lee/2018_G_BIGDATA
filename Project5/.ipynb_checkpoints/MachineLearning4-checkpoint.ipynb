## Machine Learning 4

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.metrics import classification_report, confusion_matrix
```

```python
animal_data = pd.read_csv('data_state_class.csv')
animal_data = animal_data.drop('Unnamed: 0', axis = 1)
animal_data.head()
```

```python
animal_data = animal_data[animal_data['state'] != 'X']
animal_data = animal_data[animal_data['state'] != 'PR']
```

```python
animal_data['state'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_all = animal_data.drop(['state'], axis = 1)
y_all = animal_data['state']
```

```python
X_train_A, X_final_test, y_train_A, y_final_test = train_test_split(X_all, y_all, test_size = 0.2, random_state = 101)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train_A, y_train_A, test_size = 0.2, random_state = 100)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
logistic_prediction = logistic_model.predict(X_test)

print(classification_report(y_test, logistic_prediction))
print(confusion_matrix(y_test, logistic_prediction))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors=1)
knn_model.fit(X_train, y_train)
knn_prediction = knn_model.predict(X_test)

print(classification_report(y_test, knn_prediction))
print(confusion_matrix(y_test, knn_prediction))
```

```python
knn_error_rate = []

for i in range(1, 51):
    knn_upgraded_model = KNeighborsClassifier(n_neighbors=i)
    knn_upgraded_model.fit(X_train, y_train)
    i_knn_model = knn_upgraded_model.predict(X_test)
    knn_error_rate.append(np.mean(i_knn_model != y_test))
    if i % 10 == 0:
        print('test number : ', i)
```

```python
plt.figure(figsize = (10, 5))
plt.plot(range(1, 51), knn_error_rate, marker = 'o', markerfacecolor = 'red')
plt.title('KNN Error Rate by K Value')
plt.xlabel('K Value')
plt.ylabel('Error Rate')
```

```python
knn_error_rate.index(min(knn_error_rate)) + 1
# find minimum error rate K Value
```

```python
knn_final_model = KNeighborsClassifier(n_neighbors=40)
knn_final_model.fit(X_train, y_train)
knn_final_prediction = knn_final_model.predict(X_test)

print('With K Value = 40')
print(classification_report(y_test, knn_final_prediction))
print(confusion_matrix(y_test, knn_final_prediction))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree_model = DecisionTreeClassifier()
dtree_model.fit(X_train, y_train)
dtree_prediction = dtree_model.predict(X_test)
```

```python
print(classification_report(y_test, dtree_prediction))
print(confusion_matrix(y_test, dtree_prediction))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(n_estimators=100)
rf_model.fit(X_train, y_train)
rf_prediction = rf_model.predict(X_test)
```

```python
print(classification_report(y_test, rf_prediction))
print(confusion_matrix(y_test, rf_prediction))
```

## Tensorflow Softmax

```python
import tensorflow as tf
```

```python
tf.set_random_seed(42)
```

```python
X_train_tf = X_train
y_train_tf = pd.get_dummies(y_train)
```

```python
test1 = [[1, 2, 1, 1],
          [2, 1, 3, 2],
          [3, 1, 3, 4],
          [4, 1, 5, 5],
          [1, 7, 5, 5],
          [1, 2, 5, 6],
          [1, 6, 6, 6],
          [1, 7, 7, 7]]
test1
```

```python
X_train_tf = X_train_tf.values.tolist()
y_train_tf = y_train_tf.values.tolist()
```

```python
print(type(X_train_tf))
print(type(y_train_tf))
```

```python
y_train_tf
```

```python
X_train_tf
```

```python
X = tf.placeholder("float", [None, 11])
Y = tf.placeholder("float", [None, 4])
nb_classes = 4
W = tf.Variable(tf.random_normal([11, nb_classes]), name='weight')
b = tf.Variable(tf.random_normal([nb_classes]), name='bias')
```

```python
hypothesis = tf.nn.softmax(tf.matmul(X, W) + b)
```

```python
cost = tf.reduce_mean(-tf.reduce_sum(Y * tf.log(hypothesis), axis=1))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.005).minimize(cost)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(101):
        sess.run(optimizer, feed_dict={X: X_train_tf, Y: y_train_tf})
        if step % 10 == 0:
            print(step, sess.run(cost, feed_dict = {X: X_train_tf, Y: y_train_tf}))
            
    print('-----------')
```
