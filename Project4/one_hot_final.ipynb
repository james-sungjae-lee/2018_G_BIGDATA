```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('pre_done.csv')
data.head()
```

```python
data2018 = data[(data['hapY'] == 2018)]
data2017 = data[(data['hapY'] == 2017)]
```

```python
data2018 = data2018[(data['hapM'] != 7)]
data2018 = data2018[(data['hapM'] != 6)]
```

```python
data2018['hapM'].value_counts()
```

```python
frames = [data2017, data2018]
data = pd.concat(frames)
```

```python
data = data.drop(['Unnamed: 0', 'careNm', 'color', 'care1', 'care2', 'care3', 'noticeNo', 'hapY', 'hapD', 'good', 'safe'], axis = 1)
data.head()
```

```python
data['newAge'] = data['age'].apply(lambda x: 2018 - x)
data['newAge'].value_counts()
```

```python
data = data.drop(['age'], axis = 1)
data.info()
```

```python
data.head()
```

```python
data_onehot = pd.concat([data, pd.get_dummies(data['neuter'])], axis=1, sort=False)
```

```python
data_onehot = pd.concat([data_onehot, pd.get_dummies(data['kind'])], axis=1, sort=False)
```

```python
data_onehot = pd.concat([data_onehot, pd.get_dummies(data['sexCd'])], axis=1, sort=False)
```

```python
data_onehot = data_onehot.drop(['neuter','kind'], axis = 1)
data_onehot.head()
```

```python
data_onehot = data_onehot.drop(['sexCd'], axis = 1)
```

```python
data_onehot.info()
```

```python
data_onehot = data_onehot.rename(index=str, columns={"개": "dog", "고양이": "cat", 'newAge':'age', 'N':'neuterN',
                                      'U':'neuterU', 'Y':'neuterY', 'F':'female', 'M':'male',
                                      'Q':'sexUnknown', '기타축종':'kindETC'})
data_onehot.info()
```

```python
data_onehot = data_onehot.drop(['female', 'male', 'sexUnknown'], axis = 1)
```

```python
data_onehot = data_onehot.drop(['종료(방사)', '종료(미포획)', '종료(기증)'], axis = 1)
data_onehot.head()
```

```python
data_onehot.head()
```

```python
data_onehot['items'] = data_onehot['item'].apply(lambda x: int(x))
data_onehot['bad'] = data_onehot['sad'].apply(lambda x: int(x))
data_onehot = data_onehot.drop(['sad', 'item'], axis = 1)
data_onehot.head()
```

```python
from sklearn.preprocessing import Imputer
imputer = Imputer(strategy = 'median')
imputer.fit(data_onehot.drop(['state'], axis = 1))
```

```python
data_onehot.info()
```

```python
data_onehot_part = data_onehot.drop(['state'], axis = 1)
```

```python
X = imputer.transform(data_onehot.drop(['state'], axis = 1))
data_onehot_new = pd.DataFrame(X, columns=data_onehot_part.columns, index = list(data_onehot.index.values))
data_onehot_new.head()
```

```python
data_done = pd.concat([data_onehot_new, data_onehot['state']], axis=1, sort=False)
data_done.head()
```

```python
data_done.info()
```

```python
data_done['state'].value_counts()
```

```python
data_done = data_done.replace({'종료(입양)': 'AD', '종료(자연사)': 'ND', '종료(안락사)':'ED', '종료(반환)':'RE', '보호중':'PR',
                   '종료(기증)': 'X', '종료(방사)':'X', '종료(미포획)':'X'})
```

```python
data_done['state'].value_counts()
```

```python
data_done.info()
```
