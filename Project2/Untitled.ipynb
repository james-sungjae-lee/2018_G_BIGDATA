```python
after_pre = after_pre.drop(['Unnamed: 0'], axis = 1)
after_pre.head(1)
```

```python
after_pre = pd.concat([after_pre, pd.get_dummies(after_pre['neuter'])], axis=1, sort=False)
```

```python
after_pre = pd.concat([after_pre, pd.get_dummies(after_pre['state'])], axis=1, sort=False)
```

```python
after_pre = pd.concat([after_pre, pd.get_dummies(after_pre['kind'])], axis=1, sort=False)
```
