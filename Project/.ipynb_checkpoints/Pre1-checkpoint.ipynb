```python
import pandas as pd
```

```python
age_done = pd.read_csv('age_done.csv')
```

```python
age_done.info()
```

```python
age_done = age_done.drop(['Unnamed: 0', 'Unnamed: 16'], axis = 1)
```

```python
age_done.head()
```

```python
age_done.info()
```

```python
age_image = age_done.drop(['popfile'], axis = 1)
```

```python
age_image.head()
```

```python
age_image['new'] = age_image.new.str.split('[').str[1]
```

```python
age_image_kind = age_image.drop(['kindCd'], axis = 1)
```

```python
age_image_kind.rename(columns = {'new':'kind'}, inplace = True)
```

```python
age_image_kind = age_image_kind.drop(['happenPlace'], axis = 1)
```

```python
pre1 = age_image_kind
```

```python
pre1['care5'] = pre1.careAddr.str.split(' ').str[4]
```

```python
pre1 = pre1.drop(['care5'], axis = 1)
```

```python
pre1['notice3'] = pre1.noticeNo.str.split('-').str[3]
```

```python
pre1 = pre1.drop(['noticeNo'], axis = 1)
```

```python
pre1.head()
```

```python
pre1['weight1'] = pre1.weight.str.split('(').str[0]
```

```python
pre1 = pre1.drop(['weight'], axis = 1)
```

```python
pre1.head()
```

```python
pre1.rename(columns = {'notice3':'noticeNo'}, inplace = True)
```

```python
pre1 = pre1.drop(['notice1', 'notice2'], axis = 1)
```

```python
pre1.weight.str.split('(').str[0]
```

```python
pre1['happenStr'] = pre1.happenDt.apply(lambda x: str(x))
```

```python
pre1['happenYear'] = pre1.happenStr.apply(lambda x: x[:4])
```

```python
pre1['happenMonth'] = pre1.happenStr.apply(lambda x: x[4:6])
```

```python
pre1['happenDay'] = pre1.happenStr.apply(lambda x: x[6:])
```

```python
pre1 = pre1.drop(['happenStr', 'happenDt'], axis = 1)
```

```python
pre1.to_csv('pre1.txt', encoding='utf-8')
```
